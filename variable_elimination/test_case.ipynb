```python
cpd_poll = TabularCPD(variable='P', values=[[0.9, 0.1]])
cpd_smoke = TabularCPD(variable='S', values=[[0.3, 0.7]])
cpd_cancer = TabularCPD(variable='C', values=[[0.03, 0.05, 0.001, 0.02],[0.97, 0.95, 0.999, 0.98]],evidence=['S', 'P'],)
cpd_xray = TabularCPD(variable='X', values=[[0.9, 0.2], [0.1, 0.8]],evidence=['C'])
cpd_dysp = TabularCPD(variable='D', values=[[0.65, 0.3], [0.35, 0.7]], evidence=['C'])
cancer_model = BayesModel([('P', 'C'), ('S', 'C'), ('C', 'X'), ('C', 'D')])
cancer_model.add_cpds(cpd_poll, cpd_smoke, cpd_cancer, cpd_xray, cpd_dysp)
infer3 = VarElimination(cancer_model)
# Q1
print(infer3.query(['X']))
# Q2
print(infer3.query(['C'],{'P':0,'S':0,'D':1}))
# Q3
print(infer3.query(['S'],{'D':1}))
```

```python
import numpy as np
from pgmpy.models import BayesianModel
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination
```

```python
cancer_model = BayesianModel([('Pollution', 'Cancer'), 
                              ('Smoker', 'Cancer'),
                              ('Cancer', 'Xray'),
                              ('Cancer', 'Dyspnoea')])
cpd_poll = TabularCPD(variable='Pollution', variable_card=2,
                      values=[[0.9], [0.1]])
cpd_smoke = TabularCPD(variable='Smoker', variable_card=2,
                       values=[[0.3], [0.7]])
cpd_cancer = TabularCPD(variable='Cancer', variable_card=2,
                        values=[[0.03, 0.05, 0.001, 0.02],
                                [0.97, 0.95, 0.999, 0.98]],
                        evidence=['Smoker', 'Pollution'],
                        evidence_card=[2, 2])
cpd_xray = TabularCPD(variable='Xray', variable_card=2,
                      values=[[0.9, 0.2], [0.1, 0.8]],
                      evidence=['Cancer'], evidence_card=[2])
cpd_dysp = TabularCPD(variable='Dyspnoea', variable_card=2,
                      values=[[0.65, 0.3], [0.35, 0.7]],
                      evidence=['Cancer'], evidence_card=[2])
cancer_model.add_cpds(cpd_poll, cpd_smoke, cpd_cancer, cpd_xray, cpd_dysp)
infer_t2 = VariableElimination(cancer_model)
# Q1
print(infer_t2.query(variables = ['Xray']))
# Q2
print(infer_t2.query(variables = ['Cancer'], evidence = {'Pollution':0, 'Smoker':0, 'Dyspnoea':1}))
# Q3
print(infer_t2.query(variables = ['Smoker'], evidence = {'Dyspnoea':1}))
```
